# Matplotlib

```python
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
x =  [1,2,3,4,5,6,7]
y = [40, 41, 40, 43, 42, 44, 40]

plt.plot(x, y, color='blue')
plt.scatter(x,y)
plt.xlabel('Day')
plt.ylabel('Temperature') 
plt.title("Last 7 Days Tmperature in Dhaka");
```

```python
x =  [1,2,3,4,5,6,7]
y = [40, 41, 40, 43, 42, 44, 40]

plt.plot(x,y, 'g+-.');
```

```python
days =  [1,2,3,4,5,6,7]
max_T = [40, 41, 40, 43, 42, 44, 40]
min_T = [37, 38, 34, 35, 39, 36, 38]
avg_T = [39.5, 38, 39, 37, 41, 36, 35] 

plt.plot(days, max_T, label = 'Max')
plt.plot(days, min_T, label = 'Min')
plt.plot(days, avg_T, label = 'Avarage') 
plt.legend(loc = 'lower left',shadow = True)
plt.grid() 
plt.title("Last 7 Days Tmperature Report In Dhaka");
```

# Bar Chart

```python
import numpy as np
```

```python
products = np.array([1,2,3,4])
prices = [20, 30, 35, 20]
profit = [5, 4, 10, 7]
plt.bar(products-.2, prices, label = 'price', color='red', width = .4) 
plt.bar(products+.2,profit, label='profit', color='blue',  width = .4)
plt.legend()
```

# Histogram

```python
price = [12,23,4,34,32,21,23,43,23,3,4,3,23,22,14,31,21,12,13,32,32,32,]

plt.hist(price, bins = [12, 23,])
```

```python
val = [25, 20, 10, 5, 20,16,4] 
lab = ['Bannana', 'Apple', 'Mango', 'Lamon Orange', 'Brade', 'jackfruit', 'Water']

plt.pie(val, labels = lab, radius = 2, autopct = '%.0f%%', shadow = True, startangle=45, explode = [0,0,0,0,0,0.2,0]); 
plt.savefig('piechart.png')
```
